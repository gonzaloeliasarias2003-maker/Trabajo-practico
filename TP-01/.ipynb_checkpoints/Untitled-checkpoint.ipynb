```python
btc = df[df['Symbol'] == 'BTC']
btc.head()
```

```python
btc['Date'] = pd.to_datetime(btc['Date'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))
plt.plot(btc['Date'], btc['Close'], color='blue')


plt.title('Precio de cierre de Bitcoin a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre (USD)')

plt.grid(True)
plt.show()
```

```python
btc2016 = btc[btc['Date'] >= '2016-01-01']
plt.figure(figsize=(12,7))
plt.plot(btc2016['Date'],btc2016['Close'],color='blue')

plt.title('Precio de cierre de Bitcoin a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre (USD)')

plt.grid(True)
plt.show()
```
